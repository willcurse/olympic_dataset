# 1. Which country won the most total medals?

```python
max_medal=data.loc[data['total'].idxmax(),['country','total']]
max_medal
```

# 2: Top 10 country for medal

```python
top_10=data.sort_values(by='total',ascending=False).head(10)

plt.figure(figsize=(10, 6))  
sns.barplot(x='total', y='country', data=top_10)
plt.title('Top 10 Countries by Total Medals')
plt.show()
```

# 3: Which country has the highest GDP?

```python
highest_gdp=data.loc[data['gdp'].idxmax(),['country','gdp']]
print(highest_gdp)
```

# 4: Top 10 countries..according to GDP

```python
top_10=data.sort_values(by='gdp',ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='gdp',y='country',data=top_10)
plt.title('Top 10 countries')
plt.show()
```

# 9:  What is the average number of total medals won by the countries?

```python
data.columns
```

```python
average_medal=data[['total']].mean()
print(f'The average number of total medal won by countries are: {average_medal} ')
```

# 10: Is there a correlation between GDP and the total number of medals?

```python
correlation=data['gdp'].corr(data['total'])
print(correlation)
```

# 11: What percentage of total medals did the United States win?

```python
# medal won by u.s
us_medal=data.loc[data['country']=='United States','total'].values[0]

total_medal=data['total'].sum()

percentage_of_medal=(us_medal/total_medal)*100
print(f' percentage of total medals did the United States win are {percentage_of_medal}')
```

# 12: Which country has the lowest population but still won medals?

```python
medal_winner=data[data['total']>0]
low_pop=medal_winner.loc[medal_winner['population'].idxmin()]
print(low_pop)
```

# 13: Lowest 10 Population countries?..

```python
low_pop=data.sort_values(by='population',ascending=True).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='population',y='country',data=low_pop)
plt.title(' Lowest 10 Population countries?..')
plt.show()
```

# 14: What is the total population of countries that won more than 50 medals?

```python
total_medal=data[data['total']>50]

total_pop=total_medal['population'].sum().round()
print(f'the total population of countries that won more than 50 medals are {total_pop}')
```

# 15: Top population of countries that won more than 50 medals..?
    

```python
total_medal=data[data['total']>50]
plt.figure(figsize=(10,6))
sns.barplot(x='population',y='country',data=total_medal)
plt.title('Top population of countries that won more than 50 medals..?')
plt.show()
```

# 16: Which country has the highest medals-to-population ratio?

```python
data['mean_to_population']=data['total']/data['population']
```

```python
data.columns
```

```python
highest_medal_to_pop=data.loc[data['mean_to_population'].idxmax()]
print(highest_medal_to_pop[['country', 'population', 'total', 'mean_to_population']])

```

# 17: Visualizing Medal Counts

```python
medal_types = ['gold', 'silver', 'bronze']
data[medal_types].sum().plot(kind='bar', color=['gold', 'silver', '#cd7f32'])
plt.title('Total Medal Counts by Type')
plt.ylabel('Number of Medals')
plt.show()
```

# 18: Population vs. Total Medals

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='population', y='total', hue='country', legend=False)
plt.title('Population vs. Total Medals')
plt.xlabel('Population (in millions)')
plt.ylabel('Total Medals')
plt.xscale('log')
plt.show()
```
